```python
from Bio import SeqIO

# This function extracts the accession number from a SeqRecord ID
def get_accession(record):
    parts = record.id.split("|")
    if len(parts) > 3:
        return parts[3]  # Example: 'Z78488.1'
    else:
        return record.id  # Fallback to the full ID if the expected format is not present

file_path = r"/Users/macbookairm2/Library/Mobile Documents/com~apple~CloudDocs/Biopython(own)/ls_orchid.fasta"

# Convert parsed records to a dictionary using the custom key function
orchid_dict = SeqIO.to_dict(
    SeqIO.parse(file_path, "fasta"),
    key_function=get_accession
)

print(orchid_dict.keys())
```

```python
from Bio import SeqIO

def get_acc(record_id):
    """Extract the accession number from a record ID."""
    parts = record_id.split("|")
    return parts[3] if len(parts) > 3 else record_id

def get_acc_from_record(record):
    return get_acc(record.id)

file_path = r"/Users/macbookairm2/Library/Mobile Documents/com~apple~CloudDocs/Biopython(own)/ls_orchid.fasta"

orchid_dict = SeqIO.to_dict(
    SeqIO.parse(file_path, "fasta"),
    key_function=get_acc_from_record
)

print(orchid_dict.keys())
```
