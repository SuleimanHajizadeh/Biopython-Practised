```python
import copy
original_blast_record = copy.deepcopy(blast_record)
def map_func(hit):
    # renames "gi|301171322|ref|NR_035857.1|" to "NR_035857.1"
    hit.target.id = hit.target.id.split("|")[3]
    return hit

blast_record[:] = map(map_func, blast_record)
for hit in blast_record[:5]:
    print(hit.target.id)
```
