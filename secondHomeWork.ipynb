```python
#6 Annotations from GenBank File – Extract the names of genes and their start-stop positions by reading the .gbk file.
```

```python
from Bio import SeqIO

# Enter the path to the GenBank file
file_path = r"/Users/macbookairm2/Documents/GitHub/Bioinformatics/Biopython/Data/ls_orchid.gbk"

# We read the file and extract the annotation data
for seq_record in SeqIO.parse(file_path, "genbank"):
    print("ID:", seq_record.id)
    print("Sequence Length:", len(seq_record.seq))
    
    # We iterate over each annotation (feature)
    for feature in seq_record.features:
        if feature.type == "gene":  # Take only the "gene" annotation type
            gene_name = feature.qualifiers.get("gene", ["Unknown"])[0]  # Get gene name
            start = feature.location.start
            end = feature.location.end
            print(f"Gen: {gene_name}, Start: {start}, End: {end}")

    print("-" * 50)  # Line to separate data
```

```python
#7 Find the Longest Sequence in a FASTA File – Select and print the longest sequence in the .fasta file.
```

```python
from Bio import SeqIO

# Enter the path to the FASTA file
file_path = r"/Users/macbookairm2/Documents/GitHub/Bioinformatics/Biopython/Data/ls_orchid.fasta"

# Variables to find the longest sequence
longest_seq = None
max_length = 0

# We read the file and find the longest sequence
for seq_record in SeqIO.parse(file_path, "fasta"):
    seq_length = len(seq_record.seq)
    if seq_length > max_length:  
        max_length = seq_length
        longest_seq = seq_record

# We print the longest sequence
if longest_seq:
    print("ID of the longest sequence:", longest_seq.id)
    print("Length:", max_length)
    print("Sequence:", longest_seq.seq)
    print("-" * 640)
```

```python
#9 Finding a DNA sequence motif – Check if a given DNA sequence contains a specific motif (e.g. "ATG") using the find() method.
```

```python
from Bio.Seq import Seq

# DNA sequence
my_seq = Seq("CGTAACAAGGTTTCCGTAGGTGAACCTGCGGAAGGATCATTGTTGAGACAACAGAATATATGATCGAGTG")

# Motive to search for
motif = "ATG"

# Since the find() method only works on str objects, we convert Seq to str
position = str(my_seq).find(motif)

# We print the result
if position != -1:
    print(f"Motif '{motif}' found at index {position} in the sequence.")
else:
    print(f"Motif '{motif}' not found in the sequence.")
```

```python
#10 Comparing DNA Sequences – Print out the similarities and differences by comparing two different DNA sequences.
```

```python
from Bio.Seq import Seq

# Two different DNA sequences
seq1 = Seq("CGTAACAAGGTTTCCGTAGGTGAACCTGCGGAAGGATCATTGTTGAGACAACAGAATATATGATCGAGTG")
seq2 = Seq("ACGAAGTAGGTTTCCGTAAGTGAACCTGCGGAAGGATATCGGTTGAGACAACAGAATATATGATCGAGTG")

# We compare
matching_positions = []
non_matching_positions = []

# We get the length of both sequences (they must be equal)
min_len = min(len(seq1), len(seq2))

# We compare two sequences element by element
for i in range(min_len):
    if seq1[i] == seq2[i]:
        matching_positions.append(i)
    else:
        non_matching_positions.append(i)

# We print similar and different parts
print(f"Similar parts: {matching_positions}")
print(f"Different parts: {non_matching_positions}")
```

```python
from Bio import pairwise2  # Import pairwise2 for sequence alignment

# Two different DNA sequences
seq1 = Seq("CGTAACAAGGTTTCCGTAGGTGAACCTGCGGAAGGATCATTGTTGAGACAACAGAATATATGATCGAGTG")
seq2 = Seq("ACGAAGTAGGTTTCCGTAAGTGAACCTGCGGAAGGATATCGGTTGAGACAACAGAATATATGATCGAGTG")

# To do Alignment
alignments = pairwise2.align.globalxx(seq1, seq2)

# We print the result
for alignment in alignments:
    print("Aligned sequences:")
    print("Seq1: ", alignment[0])
    print("Seq2: ", alignment[1])
    print("Score: ", alignment[2])
    print()
```

```python
#11 Find potential coding regions (regions that start with the ATG start codon and end with the TAA, TAG, or TGA stop codons) in the given DNA sequence.
```

```python
# DNA sequence
my_Seq = Seq("ATGGCATCGGTGACCAGTATGCTACGATTAAATGCGTAAGTTGCCATTATGGATCGATTCAAGTGA")

# Codones
start_codon = "ATG"
stop_codons = ["TAA", "TAG", "TGA"]

def find_coding_regions(my_Seq):
    coding_regions = []
    
    # Find the Start codones
    start_indices = [i for i in range(len(my_Seq) - 2) if my_Seq[i:i+3] == start_codon]

    for start_index in start_indices:
        possible_stops = []
        
        # Find stop codons (pick the ones that fit the reading frame)
        for stop_codon in stop_codons:
            stop_index = my_Seq.find(stop_codon, start_index + 3)
            while stop_index != -1:
                if (stop_index - start_index) % 3 == 0:  #A multiple of 3?
                    possible_stops.append(stop_index)
                    break # Get earliest stop codon
                stop_index = my_Seq.find(stop_codon, stop_index + 3)  # Look for the next stop codon
        
        # If a suitable stop codon is found, add the coding region
        if possible_stops:
            stop_index = min(possible_stops)  # Select the earliest available stop codon
            coding_regions.append(my_Seq[start_index:stop_index + 3])

    return coding_regions

# Find encoding regions
coding_regions = find_coding_regions(my_Seq)

# Print results
print("Potential coding regions found:")
for region in coding_regions:
    print(region)
```

```python
#12 Find tandem repeat motifs (i.e. short nucleotide sequences that are repeated several times in a row) in a DNA sequence and determine their number of repetitions.
```

```python
# DNA sequence
my_seq = Seq("ATATATCGCGCGCAGCTGATATATAGCGCGCGAT")

# We define the tandem motif
motif = "AT"

# We find the number of repetitions of the motif in the sequence
count = my_seq.count(motif)

# We print the result
print(f"Motif '{motif}' appears {count} times in the sequence.")
```
